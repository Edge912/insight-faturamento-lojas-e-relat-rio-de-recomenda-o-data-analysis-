# üìä An√°lise de Desempenho de Lojas de Varejo Online

Este projeto realiza uma an√°lise detalhada do desempenho de quatro lojas de varejo online, utilizando dados de vendas para identificar tend√™ncias, faturamento, categorias de produtos mais vendidas e avalia√ß√µes dos clientes. O objetivo principal √© fornecer insights acion√°veis para otimiza√ß√£o de performance e tomada de decis√µes estrat√©gicas.

## üéØ Objetivo do Projeto

Analisar o desempenho individual e comparativo de quatro lojas de varejo para:

*   Identificar as lojas com melhor e pior performance em termos de faturamento.
*   Entender a distribui√ß√£o de vendas por categoria de produto.
*   Avaliar a satisfa√ß√£o do cliente atrav√©s da m√©dia de avalia√ß√µes.
*   Analisar o frete m√©dio por loja.
*   Descobrir produtos mais e menos avaliados.
*   Visualizar a distribui√ß√£o geogr√°fica das vendas.
*   Propor recomenda√ß√µes estrat√©gicas com base nos dados.

## üîç An√°lises Realizadas

1.  **Faturamento Total por Loja**: Calculou-se o faturamento total de cada loja para comparar o desempenho financeiro.
2.  **Vendas por Categoria de Produto**: An√°lise da quantidade de vendas por categoria em cada loja e consolida√ß√£o dos dados.
3.  **M√©dia de Avalia√ß√£o das Lojas**: Calculou-se a m√©dia das avalia√ß√µes de compra para cada loja, fornecendo uma m√©trica de satisfa√ß√£o do cliente.
4.  **Produtos Mais e Menos Avaliados**: Identifica√ß√£o dos produtos com as maiores e menores m√©dias de avalia√ß√£o, tanto individualmente por loja quanto no conjunto total de dados.
5.  **Frete M√©dio por Loja**: C√°lculo do valor m√©dio do frete para cada loja.
6.  **Distribui√ß√£o Geogr√°fica das Vendas**: Visualiza√ß√£o dos pontos de venda no mapa para entender a dispers√£o geogr√°fica.

## üí° Principais Descobertas e Recomenda√ß√µes

Com base nas an√°lises, a **Loja 4** foi identificada como a unidade de **pior desempenho geral**, apresentando:

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja1.head(5)

```

#1. An√°lise do faturamento



```python
df1 = loja1
df2 = loja2
df3 = loja3
df4 = loja4

lojas = [df1, df2, df3, df4]

faturamentos = []

for i, df in enumerate(lojas, start=1):
    total = df["Pre√ßo"].sum()
    faturamentos.append([f"Loja_{i}", total])

df_faturamento = pd.DataFrame(faturamentos, columns=["Loja", "Faturamento_Total"])

df_faturamento

```

# 2. Vendas por Categoria

```python
lojas = [df1, df2, df3, df4]

lista_categorias = []

for i, df in enumerate(lojas, start=1):
    temp = (
        df.groupby("Categoria do Produto")
          .size()
          .reset_index(name=f"Loja_{i}")
    )
    lista_categorias.append(temp)

df_final = lista_categorias[0]

for temp in lista_categorias[1:]:
    df_final = df_final.merge(temp, on="Categoria do Produto", how="outer")

df_final
```

# 3. M√©dia de Avalia√ß√£o das Lojas


```python
lojas = [df1, df2, df3, df4]

medias_lojas = []

for i, df in enumerate(lojas, start=1):
    media = df["Avalia√ß√£o da compra"].mean()
    medias_lojas.append([f"Loja_{i}", media])

df_media_por_loja = pd.DataFrame(medias_lojas, columns=["Loja", "Media_Avaliacao"])
df_media_por_loja["Media_Avaliacao"] = df_media_por_loja["Media_Avaliacao"].round(2)
df_media_por_loja
```

# 4. Produtos Mais e Menos Vendidos

```python
df_todas = pd.concat([df1, df2, df3, df4])

df_media_categoria = (
    df_todas.groupby("Produto")["Avalia√ß√£o da compra"]
    .mean().round(2)
    .reset_index()
    .sort_values(by="Avalia√ß√£o da compra", ascending=False)
)

df_media_categoria_loja1 = (
    df1.groupby("Produto")["Avalia√ß√£o da compra"]
    .mean().round(2)
    .reset_index()
    .sort_values(by="Avalia√ß√£o da compra", ascending=False)
)

df_media_categoria_loja2 = (
    df2.groupby("Produto")["Avalia√ß√£o da compra"]
    .mean().round(2)
    .reset_index()
    .sort_values(by="Avalia√ß√£o da compra", ascending=False)
)

df_media_categoria_loja3 = (
    df3.groupby("Produto")["Avalia√ß√£o da compra"]
    .mean().round(2)
    .reset_index()
    .sort_values(by="Avalia√ß√£o da compra", ascending=False)
)

df_media_categoria_loja4 = (
    df4.groupby("Produto")["Avalia√ß√£o da compra"]
    .mean().round(2)
    .reset_index()
    .sort_values(by="Avalia√ß√£o da compra", ascending=False)
)

df_media_categoria_loja4
```

# 5. Frete M√©dio por Loja

```python
lojas = [df1, df2, df3, df4]

frete_medio_lojas = []

for i, df in enumerate(lojas, start=1):
    media = df["Frete"].mean()
    frete_medio_lojas.append([f"Loja_{i}", media])

df_frete_medio_por_loja = pd.DataFrame(frete_medio_lojas, columns=["Loja", "Frete"])
df_frete_medio_por_loja["Frete"] = df_frete_medio_por_loja["Frete"].round(2)
df_frete_medio_por_loja
```

# 6. Gr√°ficos de An√°lise

## Relat√≥rio Resumido de Recomenda√ß√£o de Encerramento

A an√°lise dos dados das quatro lojas ‚Äî incluindo faturamento, avalia√ß√µes dos clientes, desempenho por categoria e ranking de produtos ‚Äî evidencia diferen√ßas significativas na efici√™ncia operacional de cada unidade. Entre elas, a Loja 4 apresenta o pior desempenho geral, tanto em gera√ß√£o de receita quanto em satisfa√ß√£o do cliente.


O faturamento da Loja 4 √© inferior ao das demais, indicando baixa capacidade de contribui√ß√£o para o resultado consolidado. Al√©m disso, sua m√©dia de avalia√ß√£o √© consistentemente mais baixa, sugerindo problemas de opera√ß√£o, atendimento ou qualidade percebida. A loja tamb√©m n√£o se destaca em nenhuma categoria relevante e apresenta desempenho fraco no conjunto de produtos avaliados.


Do ponto de vista financeiro, manter uma unidade com baixa performance reduz a rentabilidade global e consome recursos que poderiam ser direcionados para lojas mais eficientes. A l√≥gica aplicada segue o princ√≠pio de maximiza√ß√£o de valor: descontinuar opera√ß√µes deficit√°rias para fortalecer √°reas mais produtivas.


Diante disso, recomenda-se o encerramento da Loja 4, por ser a unidade menos estrat√©gica e com menor potencial de recupera√ß√£o. A medida permite otimizar recursos, melhorar o desempenho consolidado e concentrar esfor√ßos nas lojas com maior retorno.

```python
import folium

df_group = df_todas.groupby(['lat', 'lon']).size().reset_index(name='qtd')

max_qtd = df_group['qtd'].max()

m = folium.Map(
    location=[df_group['lat'].mean(), df_group['lon'].mean()],
    zoom_start=12
)

for _, row in df_group.iterrows():
    raio = (row['qtd'] / max_qtd) * 25

    folium.CircleMarker(
        location=[row['lat'], row['lon']],
        radius=raio,
        color='orange',
        fill=True,
        fill_opacity=0.8
    ).add_to(m)

m
```
